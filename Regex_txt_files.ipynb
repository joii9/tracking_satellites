```python
import requests
from bs4 import BeautifulSoup as bs
from requests.packages.urllib3.exceptions import InsecureRequestWarning

requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

def getting_tle(i):

    try:
        page = requests.get('https://www.n2yo.com/satellite/?s='+str(i)+'#results' , verify=False)
        soup = bs(page.text, 'html.parser')
        pre_tags = soup.find("pre")
        str_pre_tags= str(pre_tags)
        #print(str_pre_tags[7:-7])
        if len(str_pre_tags[7:-7]) > 15:
            textfile= open("C:\\Users\\Joel\\Documents\\Python\\email\\test.txt", "a")
            textfile.write(str_pre_tags[7:-7]+"\n\n\n")
            textfile.close()
    except requests.exceptions.SSLError as e:
        print(f"SSL Error: {e}")
```
